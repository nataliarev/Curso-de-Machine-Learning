## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """select distinct LaunchSite from SPACEXTBL""" #Query contiene la consulta SQL
df  = pd.read_sql(query, con = engine )  # Con pd.read_sql se ejecuta la consulta sql por query y se almacena en df. 'con = engine' especifica la conexión de base de datos para ejecutar la consulta.
df   # El df contendrá los resultados de la consulta, que en este caso serán valores únicos de LaunchSite.
# Se ejecuta la consulta SQL en la base de datos vinculada a almacenar los resultados en Df de Pandas
# Los resultados de la consuta se reflejarán en el Dataframe df que luego se pueden utilizar para análisis de datos.
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variable query con tu consulta
query =  """select launchsite
from SPACEXTBL
WHERE launchsite LIKE 'CCS%' LIMIT 5"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """select SUM(PayloadMass), customers
from SPACEXTBL
WHERE Customers = 'NASA (CRS)'"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """select AVG(PayloadMass), serial
from SPACEXTBL
WHERE serial LIKE 'B1%'"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """ select MIN(date), LandingOutcome
from SPACEXTBL
WHERE LandingOutcome = 'TRUE RTLS' """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ select serial
from SPACEXTBL
WHERE LandingOutcome = 'TRUE ASDS' AND PayloadMass
> 4000 AND PayloadMass < 6000"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """ select MISSION_OUTCOME, COUNT(*) AS num_lanzamientos
from SPACEXTBL
GROUP BY Mission_outcome;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = """ select serial, MAX(PayloadMass) AS series_max
from SPACEXTBL
GROUP BY serial;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """ select serial, date, LandingOutcome, COUNT(*) Results_Aterrizaje
from SPACEXTBL
WHERE date = 2015
GROUP BY LandingOutcome"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query = """ select LandingOutcome, COUNT(*) AS Results_aterrizaje
from SPACEXTBL
WHERE date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY LandingOutcome
ORDER BY Results_aterrizaje DESC"""
df  = pd.read_sql(query, con = engine )
df
```
